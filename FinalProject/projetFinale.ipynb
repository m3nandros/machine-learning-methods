# 2. Описание данных

Представленный набор данных содержит информацию о том, как часто человек употребляет различные наркотики. База данных содержит записи 1885 респондентов с подробной метаинформацией.

Для каждого респондента известно 12 атрибутов: 
1. Измерения личности, включающие NEO-FFI-R (нейротизм, экстраверсия, открытость опыту, соглашательство и добросовестность);
2. BIS-11 (импульсивность) и ImpSS (стремление к ощущениям); 
3. Уровень образования; 
4. Возраст; 
5. Пол; 
6. Страна проживания;
6. Этническая принадлежность. 

Все входные признаки изначально являются категориальными и оцениваются количественно. После квантификации значения всех входных признаков могут рассматриваться как реально-значимые. 

Кроме того, участников опросили на предмет употребления 18 легальных и нелегальных наркотиков (алкоголь, амфетамины, амилнитрит, бензодиазепин, каннабис, шоколад, кокаин, кофеин, крэк, экстази, героин, кетамин, легальные наркотики, ЛСД, метадон, грибы, никотин и летучие вещества) и одного вымышленного наркотика (семирон), который был введен для выявления лиц, предъявляющих завышенные требования. Для каждого наркотика нужно выбрать один из вариантов ответа: никогда не употреблял, употреблял более десяти лет назад или за последнее десятилетие, год, месяц, неделю или день. База данных содержит 18 задач классификации. Каждая из независимых переменных-меток содержит семь классов: "никогда не употреблял", "употреблял более десяти лет назад", "употреблял в последнее десятилетие", "употреблял в последний год", "употреблял в последний месяц", "употреблял в последнюю неделю" и "употреблял в последний день".
