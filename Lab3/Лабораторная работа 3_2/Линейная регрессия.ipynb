### Проблема первая: коллинеарные признаки

Итак, в наших данных один признак дублирует другой, и есть еще два очень похожих. Конечно, мы могли бы сразу удалить дубликаты, но давайте посмотрим, как бы происходило обучение модели, если бы мы не заметили эту проблему.

Для начала проведем масштабирование, или стандартизацию признаков: из каждого признака вычтем его среднее и поделим на стандартное отклонение. Это можно сделать с помощью метода scale.

Кроме того, нужно перемешать выборку, это потребуется для кросс-валидации.



```python
from sklearn.preprocessing import scale
from sklearn.utils import shuffle
```

```python
df_shuffled = shuffle(df, random_state=123)
X = scale(df_shuffled[df_shuffled.columns[:-1]]) 
y = df_shuffled["cnt"]
```

Давайте обучим линейную регрессию на наших данных и посмотрим на веса признаков.

```python
from sklearn.linear_model import LinearRegression
```

```python
# Код 2.1 
# Создайте объект линейного регрессора, обучите его на всех данных и выведите веса модели
# (веса хранятся в переменной coef_ класса регрессора).
# Можно выводить пары (название признака, вес), воспользовавшись функцией zip, встроенной в язык python
# Названия признаков хранятся в переменной df.columns
lr_model = LinearRegression()
lr_model.fit(X,y)

print('Модель линейной регрессии')
print('Веса признаков:')
for item in zip(df_shuffled.columns[:-1], lr_model.coef_.round()):
    print(item)
```

Мы видим, что веса при линейно-зависимых признаках по модулю значительно больше, чем при других признаках.


С парой temp-atemp чуть менее коррелирующих переменных такого не произошло, однако на практике всегда стоит внимательно следить за коэффициентами при похожих признаках.

Решение проблемы мультиколлинеарности состоит в регуляризации линейной модели. К оптимизируемому функционалу прибавляют L1 или L2 норму весов, умноженную на коэффициент регуляризации α. В первом случае метод называется Lasso, а во втором --- Ridge. Подробнее об этом также рассказано в лекции.

Обучите регрессоры Ridge и Lasso с параметрами по умолчанию и убедитесь, что проблема с весами решилась.



```python
from sklearn.linear_model import Lasso, Ridge
```

```python
# Код 2.2 
# Обучите линейную модель с L1-регуляризацией
ls_model = Lasso()
ls_model.fit(X,y)
print('L1 регуляризация')
print('Веса признаков:')
for item in zip(df_shuffled.columns[:-1], ls_model.coef_.round()):
    print(item)
```

```python
# Код 2.3 
# Обучите линейную модель с L2-регуляризацией
rd_model = Ridge()
rd_model.fit(X,y)
print('L2 регуляризация')
print('Веса признаков:')
for item in zip(df_shuffled.columns[:-1], rd_model.coef_.round()):
    print(item)
```

Таким образом, избегается слишком сильное «подстраивание» под данные, что в дальнейшем улучшает спосбность модели обобщаться на новые данные.




#### Проблема вторая: неинформативные признаки

В отличие от L2-регуляризации, L1 обнуляет веса при некоторых признаках. Объяснение данному факту дается в одной из лекций курса.

Давайте пронаблюдаем, как меняются веса при увеличении коэффициента регуляризации α (в лекции коэффициент при регуляризаторе мог быть обозначен другой буквой).


```python
# Код 3.1 
alphas = np.arange(1, 500, 50)
coefs_lasso = np.zeros((alphas.shape[0], X.shape[1])) # матрица весов размера (число регрессоров) x (число признаков)
coefs_ridge = np.zeros((alphas.shape[0], X.shape[1]))
# Для каждого значения коэффициента из alphas обучите регрессор Lasso
# и запишите веса в соответствующую строку матрицы coefs_lasso (вспомните встроенную в python функцию enumerate),
# а затем обучите Ridge и запишите веса в coefs_ridge.

for idx, alpha in enumerate(alphas):
    lasso = Lasso(alpha=alpha)
    lasso.fit(X, y)
    coefs_lasso[idx, :] = lasso.coef_ 

    ridge = Ridge(alpha=alpha)
    ridge.fit(X, y)
    coefs_ridge[idx, :] = ridge.coef_
```

Визуализируем динамику весов при увеличении параметра регуляризации:



```python
# График весов Lasso
plt.subplot(1, 2, 1)
for i in range(coefs_lasso.shape[1]):
    plt.plot(alphas, coefs_lasso[:, i], label=f'Признак {i + 1}')
plt.title('Динамика весов Lasso')
plt.xlabel('Alpha (параметр регуляризации)')
plt.ylabel('Вес')
plt.xscale('log')
plt.axhline(0, color='grey', lw=0.5, ls='--')
plt.grid()

# График весов Ridge
plt.subplot(1, 2, 2)
for i in range(coefs_ridge.shape[1]):
    plt.plot(alphas, coefs_ridge[:, i], label=f'Признак {i + 1}')
plt.title('Динамика весов Ridge')
plt.xlabel('Alpha (параметр регуляризации)')
plt.ylabel('Вес')
plt.xscale('log')
plt.axhline(0, color='grey', lw=0.5, ls='--')
plt.grid()

plt.tight_layout()
plt.show()
```

2.	Что произойдет с весами Lasso, если alpha сделать очень большим? Поясните, почему так происходит.

**Ответ:** Lasso минимизирует функцию потерь, которая включает как MSE, так и штраф за велиичну коэффициентов. Штраф представляет собой $l_1$-норму (то есть сумму абсолютных значений коэффициентов): $$\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \alpha \sum_{j=1}^{p} |w_j|,$$ где $n$ — кол-во наблюдений, $y_i$ — ист. значения, $\cap{y_i}$ — предсказанные значения и $w_j$ — коэффициенты модели. При увеличении $\alpha$ штраф возрастает, Lasso стремится уменьшить значения всех коэффициентов, что в конечном итоге схлопывают их все к нулю. Как следствие, признаки с высокими значениями коэффициентов при малых $\alpha$ могут оказаться незначительными и будут обнулевы при больших значениях $\alpha$. Так Lasso проводит отбор признаков.

###### Далее будем работать с Lasso.

Итак, мы видим, что при изменении alpha модель по-разному подбирает коэффициенты признаков. Нам нужно выбрать наилучшее alpha.

Для этого, во-первых, нам нужна метрика качества. Будем использовать в качестве метрики сам оптимизируемый функционал метода наименьших квадратов, то есть Mean Square Error.

Во-вторых, нужно понять, на каких данных эту метрику считать. Нельзя выбирать alpha по значению MSE на обучающей выборке, потому что тогда мы не сможем оценить, как модель будет делать предсказания на новых для нее данных. Если мы выберем одно разбиение выборки на обучающую и тестовую (это называется holdout), то настроимся на конкретные "новые" данные, и вновь можем переобучиться. Поэтому будем делать несколько разбиений выборки, на каждом пробовать разные значения alpha, а затем усреднять MSE. Удобнее всего делать такие разбиения кросс-валидацией, то есть разделить выборку на K частей, или блоков, и каждый раз брать одну из них как тестовую, а из оставшихся блоков составлять обучающую выборку.

Делать кросс-валидацию для регрессии в sklearn совсем просто: для этого есть специальный регрессор, LassoCV, который берет на вход список из alpha и для каждого из них вычисляет MSE на кросс-валидации. После обучения (если оставить параметр cv=3 по умолчанию) регрессор будет содержать переменную mse_path_, матрицу размера len(alpha) x k, k = 3 (число блоков в кросс- валидации), содержащую значения MSE на тесте для соответствующих запусков. Кроме того, в переменной alpha_ будет храниться выбранное значение параметра регуляризации, а в coef_, традиционно, обученные веса, соответствующие этому alpha_.

Обратите внимание, что регрессор может менять порядок, в котором он проходит по alphas; для сопоставления с матрицей MSE лучше использовать переменную регрессора alphas_.



```python
from sklearn.linear_model import LassoCV
```

```python
# Код 3.2 
# Обучите регрессор LassoCV на всех параметрах регуляризации из alpha
# Постройте график _усредненного_ по строкам MSE в зависимости от alpha.
# Выведите выбранное alpha, а также пары "признак-коэффициент" для обученного вектора коэффициентов
alphas = np.arange(1, 100, 5)

from sklearn.linear_model import LassoCV
from sklearn.metrics import mean_squared_error

# Обучение LassoCV
lasso_cv = LassoCV(alphas=alphas, cv=5)  # 5-кратная кросс-валидация
lasso_cv.fit(X, y)

# Предсказания на обучающей выборке
y_pred = lasso_cv.predict(X)

# Среднеквадратичная ошибка (MSE)
mse = mean_squared_error(y, y_pred)
```

```python
# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(alphas, lasso_cv.mse_path_, '-o', label='MSE по каждому alpha', markersize=5)
plt.axvline(lasso_cv.alpha_, color='red', linestyle='--', label='Выбранное alpha')
plt.title('MSE в зависимости от alpha')
plt.xlabel('Alpha (параметр регуляризации)')
plt.ylabel('MSE')
plt.xscale('linear')
plt.grid()
plt.legend()
plt.show()
```

```python
# Вывод выбранного alpha
print(f'Выбранное alpha: {lasso_cv.alpha_}')

# Пары "признак-коэффициент" для обученного вектора коэффициентов
coef = lasso_cv.coef_
features = df.columns  # Получение названий признаков из DataFrame

# Вывод пар "признак-коэффициент"
for feature, weight in zip(features, coef):
    print(f'{feature}: {weight}')
```

Итак, мы выбрали некоторый параметр регуляризации. Давайте посмотрим, какие бы мы выбирали                                   alpha, если бы делили выборку только один раз на обучающую и тестовую, то есть рассмотрим траектории MSE, соответствующие отдельным блокам выборки.

```python
# Код 3.3 
# Выведите значения alpha, соответствующие минимумам MSE на каждом разбиении (то есть п о столбцам).
# На трех отдельных графиках визуализируйте столбцы .mse_path_

# Получаем значения alpha, соответствующие минимумам MSE
alpha_mins = lasso_cv.alphas_[np.argmin(lasso_cv.mse_path_, axis=1)]

# Печатаем значения alpha
print("Значения alpha, соответствующие минимумам MSE на каждом разбиении:")
print(alpha_mins)
```

На каждом разбиении оптимальное значение alpha свое, и ему соответствует большое MSE на других разбиениях. Получается, что мы настраиваемся на конкретные обучающие и контрольные выборки.
При выборе alpha на кросс-валидации мы выбираем нечто "среднее", что будет давать приемлемое значение метрики на разных разбиениях выборки.

**Ответ:** windspeed(ms). Модель исключила этот признак по той причине, что не вносит значительного вклада в объяснение зависимости между признаками и целевой переменной. Нельзя сказать, что он никак не влияет на велосипеды, но однозначно можно утверждать, что он малозначим, а значит может быть игнорирован.


```python
# Получаем коэффициенты модели
coefficients = lasso_cv.coef_

# Получение индексов 4 признаков с наибольшими положительными коэффициентами
top_positive_indices = np.argsort(coefficients)[-4:]

# Получение названий признаков
feature_names = df.columns  # Предполагаем, что X - DataFrame
top_positive_features = feature_names[top_positive_indices]

# Печатаем 4 признака с наибольшими положительными коэффициентами
print("4 признака с наибольшими положительными коэффициентами:")
for feature in top_positive_features:
    print(f"Признак: {feature}, коэффициент: {coefficients[feature_names.get_loc(feature)]}")
```

```python
top_negative_indices = np.argsort(coefficients)[:4]

top_negative_features = feature_names[top_negative_indices]

# Печатаем 4 признака с наибольшими положительными коэффициентами
print("4 признака с наибольшими по модулю отрицательными коэффициентами:")
for feature in top_negative_features:
    print(f"Признак: {feature}, коэффициент: {coefficients[feature_names.get_loc(feature)]}")
```

```python
# Индексы признаков с коэффициентами близкими к нулю
low_coef_indices = np.where(np.abs(coefficients) < np.exp(-3))[0]

# Печатаем названия признаков и их коэффициенты
print("Признаки с коэффициентами < 1e-3:")
for index in low_coef_indices:
    print(f"{feature}: коэффициент = {coefficients[index]}")
```
